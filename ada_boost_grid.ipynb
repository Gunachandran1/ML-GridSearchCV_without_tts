```python
import pandas as pd
dataset = pd.read_csv("insurance_pre.csv")
dataset.head()
```

```python
dataset= pd.get_dummies(dataset,drop_first=True,dtype=int)
dataset.head()
```

```python
features=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
target=dataset[['charges']]
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import GridSearchCV
param={'n_estimators':[50,100,200],'loss':['linear','square','exponential']}
grid=GridSearchCV(AdaBoostRegressor(),param,refit=True,verbose=3,n_jobs=-1)
grid.fit(features,target)
```

```python
grid.best_params_
```

```python
re=grid.cv_results_
data=pd.DataFrame.from_dict(re)
data
```

```python
y_pred=grid.predict(features)
y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(target,y_pred)
r_score
```
