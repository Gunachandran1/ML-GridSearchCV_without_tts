```python

import pandas as pd
dataset= pd.read_csv("insurance_pre.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
dataset.head()
```

```python
target=dataset.iloc[:,[3]]
```

```python
target
```

```python
dataset.columns
```

```python
features=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
features
```

```python
from sklearn.linear_model import QuantileRegressor
from sklearn.model_selection import GridSearchCV
params={
    'alpha': [0.01, 0.1, 1, 10, 100],
'quantile': [0.25, 0.5, 0.75]}
grid=GridSearchCV(QuantileRegressor(),params,refit=True,verbose=3,n_jobs=-1)
grid.fit(features,target)
```

```python
grid.best_params_
```

```python
re=grid.cv_results_
data=pd.DataFrame.from_dict(re)
data
```

```python
y_pred=grid.predict(features)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_pred,target)
```
